```python
import requests
```

```python
url = 'https://www.bundestag.de/ajax/filterlist/de/parlament/plenum/abstimmung/liste/462112-462112'

params = dict(
    enddate='1732834800000',
    endfield='date',
    limit='30',
    startdate='946681200000',
    startfield='date',
    noFilterSet='false',
)

resp = requests.get(url=url, params=params)
```

```python
resp.content
```

```python
xlsx_documents = []
parts = resp.content.decode("utf-8").split('"')
for part in parts:
    if '.xlsx' in part:
        xlsx_documents.append(part)

print(len(xlsx_documents))
xlsx_documents
```

```python
titles = []
parts = resp.content.decode("utf-8").split('strong>')
for part in parts[1:]:
    if not part.startswith('</'):
        part = part.replace('<span>','')
        part = part.replace('</span>','')
        part = part.replace('</','')
        part = part.replace('\n',' ')
        part = part.split(':')[-1]
        titles.append(part)

titles
```

```python
dates = []
for document in xlsx_documents:
    file_name = document.split('/')[-1]
    date = file_name.split('_')[0]
    day = date[-2:]
    month = date[-4:-2]
    year = date[:-4]
    dates.append(day + '.' + month + '.' + year)

dates
```

```python
laws = []
for i in range(len(xlsx_documents)):
    laws.append({
        'date' : dates[i],
        'title' : titles[i],
        'docuemnt' : xlsx_documents[i] 
    })

laws
```

```python
for document in xlsx_documents:
    url = 'https://www.bundestag.de' + document

    params = dict()

    resp = requests.get(url=url, params=params)

    file_name = document.split('/')[-1]
    with open(f"documents/{file_name}", "xb") as file:
        file.write(resp.content)
    
```
